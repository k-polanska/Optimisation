```python
import parser
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
def parse():
    #formula = "sin(x)*x**2"
    formula = input("enter f(x): ")
    a = int(input("enter interval. a = "))
    b = int(input("b = "))

    code = parser.expr(formula).compile()

    return code, a, b
```

```python
def goldenSection(code, a, b, epsilon=0.1):

    alpha = 0.618       # correction coefficient

    # step 0
    L = a + (1 - alpha)*(b - a)
    N = a + alpha*(b - a)
    A = a
    B = b
    n = 1

    while 1:
        # step 1
        if B - A < epsilon:
            x = (A + B) / 2
            #print(eval(code), ", A = ", A, ", B = ", B)
            return n
        else:
            x = L
            y1 = eval(code)

            x = N
            y2 = eval(code)

            if y1 > y2:
                # step 2
                A = L
                L = N
                N = A + alpha*(B - A)
            else:
                # step 3
                B = N
                N = L
                L = A + (1 - alpha) * (B - A)
            n = n+1
```

### Relationship between the minimum search interval [a,b] and the number of iterations n to find a solution

```python
formula = "sin(x)*x**2"
code = parser.expr(formula).compile()

n = []
ab = []

a = 0

for b in range(1, 100):
    n.append(goldenSection(code, a ,b))
    ab.append(b-a)
    
plt.figure(num=None, figsize=(12, 6), dpi=800, facecolor='w', edgecolor='k')
plt.plot(ab, n)
plt.title("sin(x)*x**2")
plt.xlabel('[a, b]')
plt.ylabel('n')
plt.grid()
plt.show()
```

### Relationship between the   tolerance E>0 and the  number of iterations n, required by the algorithm to  find a solution 

```python
formula = "sin(x)*x**2"
code = parser.expr(formula).compile()

n = []
E = []

a = 0
b = 10

for e in range(10000, 1, -1):
    ee = e*0.0001
    n.append(goldenSection(code, a ,b, epsilon=ee))
    E.append(ee)
    
plt.figure(num=None, figsize=(12, 6), dpi=800, facecolor='w', edgecolor='k')
plt.plot(E, n)
plt.title("sin(x)*x**2")
plt.xlabel('E')
plt.ylabel('n')
plt.grid()
plt.show()
```
